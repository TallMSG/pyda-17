# Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
best_movies = ((ratings[ratings['rating'] == 5.0])
    .merge(movies, on='movieId', how='left')
    .groupby(['movieId', 'title', 'genres'])
    .size()
    .reset_index(name='counts')
    .sort_values('counts', ascending=False))
best_movies
```

```python
print(f"Больше всего оценок 5.0 было выставлено фильму {best_movies['title'].iloc[0]}: {best_movies['counts'].iloc[0]}")
```

# Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
power_request = (data[((data['country']=='Lithuania') | (data['country']=='Latvia') | (data['country']=='Estonia')) & ((data['category']== 4) | (data['category']== 12) | (data['category']== 21)) & ((data['year'] > 2005) & (data['year'] < 2010)) & (data['quantity'] >= 0)])
print(f"Суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года составляет {power_request['quantity'].sum()}")
```

# Задание 3

Выберите страницу любого сайта с табличными данными. Импортируйте таблицы в pandas dataframe.

```python
# Определяем прирост численности населения городов России и находим самый быстрорастущий город.
# Исходные данные из Википедии.

page_url = 'https://ru.wikipedia.org/wiki/%D0%A1%D0%BF%D0%B8%D1%81%D0%BE%D0%BA_%D0%B3%D0%BE%D1%80%D0%BE%D0%B4%D0%BE%D0%B2_%D0%A0%D0%BE%D1%81%D1%81%D0%B8%D0%B8_%D1%81_%D0%BD%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5%D0%BC_%D0%B1%D0%BE%D0%BB%D0%B5%D0%B5_100_%D1%82%D1%8B%D1%81%D1%8F%D1%87_%D0%B6%D0%B8%D1%82%D0%B5%D0%BB%D0%B5%D0%B9'
df = pd.read_html(page_url, header=1)[0]
df
```

```python
df.info()
```

```python
df['pop_growth_%'] = round((df['2020'] * 100 / df['2011'] - 100), 2)
df
```

```python
df_upd = df.sort_values(by='pop_growth_%', ascending=False)
df_upd
```

```python
print(f"Самый быстрорастущий город России - {df_upd['Город'].iloc[0]}, рост населения в 2020 году составил {df_upd['pop_growth_%'].iloc[0]}% к уровню 2011 года. Город с максимальной убылью населения в России - {df_upd['Город'].iloc[-1]}, убыль населения в 2020 году составила {df_upd['pop_growth_%'].iloc[-1]}% к уровню 2011 года")
```
