# Задание 2

В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
def outliers(param):
    q1 = df[param].quantile(0.25)
    q3 = df[param].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - (1.5 * iqr) 
    upper_bound = q3 + (1.5 * iqr)
    remove_outliers = df[df[param].between(lower_bound, upper_bound, inclusive=True)].sort_values(param)
    outliers = pd.concat([df, remove_outliers]).drop_duplicates(keep=False).sort_values(param)
    return remove_outliers
```

```python
outliers('Total protein')
```

Анализ непрерывных количественных показателей ректальной температуры, пульса и частоты дыхания у лошадей показал их значительную вариабельность. В то же время произведённый анализ выбросов по указанным показателям не выявил возможных ошибок в исходных данных. Все показатели, которые были идентифицированы как выбросы по критерию межквартильного размаха в реальности могут встечаться (исходя из поверхностной проработки вопроса по материалам Интернета). Остальные анализируемые показатели вообще не дали выбросов из всей совокупности.  
Исходя из результатов анализа было принято решение не исключать никакие образцы из изначального набора данных.
