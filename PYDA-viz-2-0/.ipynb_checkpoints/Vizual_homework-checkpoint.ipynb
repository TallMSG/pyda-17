```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 13,6
```

```python
df_2015 = pd.read_csv('archive/2015.csv')
df_2015.head()
```

```python
df_2016 = pd.read_csv('archive/2016.csv')
df_2016.head()
```

```python
df_2017 = pd.read_csv('archive/2017.csv')
df_2017.head()
```

```python
df_2018 = pd.read_csv('archive/2018.csv')
df_2018.head()
```

```python
df_2019 = pd.read_csv('archive/2019.csv')
df_2019.head()
```

```python
df_2019.sort_values('Perceptions of corruption', ascending=False)
```

```python
df_2019.info()
```

```python
happy = df_2019.head(10)
unhappy = df_2019.sort_values('Score', ascending=True).head(10)
happy_unhappy = happy.append(unhappy).sort_values('Score', ascending=False)
happy_unhappy
```

```python
# Уровень счастья 10 самых счастливых и 10 самых несчастных стран мира

happy_unhappy_chart = happy_unhappy.set_index('Country or region')['Score'].plot(
    kind='bar',
    figsize=(12,8),
    title='Уровень счастья 10 самых счастливых и 10 самых несчастных стран мира в 2019 году'
)

happy_unhappy_chart.set_xlabel('Страны')
happy_unhappy_chart.set_ylabel('Индекс счастья')
```

В общем-то всё закономерно: впереди Северная Европа, позади Африка

```python
region_hapiness = df_2016.groupby('Region')[['Happiness Score']].mean().sort_values('Happiness Score', ascending=False).reset_index()
region_hapiness
```

```python
# Уровень счастья в разрезе макрорегионов мира

reg_happy_chart = region_hapiness.set_index('Region')['Happiness Score'].plot(
    kind='bar',
    figsize=(12,8),
    title='Самые счастливые и несчастные макрорегионы мира в 2016 году'
)
reg_happy_chart.set_xlabel('Макрорегионы мира')
reg_happy_chart.set_ylabel('Индекс счастья')
```

```python
df_2015_light = df_2015[['Country', 'Happiness Score']].rename({'Happiness Score': 'Score'}, axis=1)
df_2015_light.describe()
```

```python
df_2019_light = df_2019[['Country or region', 'Score']].rename({'Country or region': 'Country'}, axis=1)
df_2019_light.describe()
```

```python
# Гистограмма распределения уровня счастья сравнение 2015 и 2019

bins = np.linspace(2, 8, 7)

plt.hist([df_2015_light['Score'], df_2019_light['Score']], bins, label=['2015', '2019'])
plt.legend(loc='upper right')
plt.xlabel('Оценки индекса счастья')
plt.ylabel('Частота')
# plt.title(label='Распределение индекса счастья в 2015 и 2019 годах') - заголовок почему-то не вставляется
plt.show()
```

Все показатели говорят о том что мир за 5 лет стал чуть более счастливым

```python
# Взаимосвязь уровня счастья с различными факторами - скатерограммы

# Проверим зависимость уровня счастья от ВВП на душу населения
ax = sns.regplot(data=df_2019, x='Score', y='GDP per capita', color='green', marker='+')
ax.set_title('Зависимость уровня счастья от ВВП на душу населения')
```

Очевидно, зависимость прослеживается. Интересно было бы найти самую богатую и несчастную страну и самую счастливую и бедную страну

```python
# Выясним основные факторы, влияющие на уровень счастья

sns.pairplot(df_2019[['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption' ]])
```

Похоже что высокий уровень ВВП всё-таки является главным критерием счастья

```python
# Динамика уровня счастья с 2015 по 2019 годы в России и странах-соседях графики

Russia_neighbour = df_2019.loc[df_2019['Country or region'].isin(['Russia','Belarus', 'Ukraine', 'Kazakhstan', 'Finland', 'Poland', 'Turkey', 'Iran', 'Georgia', 'China', 'Japan', 'United States'])]
Russia_neighbour
```

```python
Russia_neighbour = Russia_neighbour.merge(df_2018, how='left', on='Country or region').rename({'Country or region': 'Country'}, axis=1).merge(df_2017, how='left', on='Country').merge(df_2016, how='left', on='Country').merge(df_2015, how='left', on='Country')
```

```python
Russia_neighbour.info()
```

```python
Russia_neighbour_hapiness_score = Russia_neighbour[['Country', 'Happiness Score_y','Happiness Score_x', 'Happiness.Score', 'Score_y', 'Score_x']].rename({'Happiness Score_y': '2015','Happiness Score_x': '2016', 'Happiness.Score': '2017', 'Score_y': '2018', 'Score_x': '2019'}, axis=1)
Russia_neighbour_hapiness_score = Russia_neighbour_hapiness_score.set_index('Country')
Russia_neighbour_hapiness_score
```

```python
t_df = Russia_neighbour_hapiness_score.T
t_df
```

```python
for el in t_df.columns:
    plt.plot(t_df.index, t_df[el])
    plt.title('Динамика индекса счастья по России и отдельным соседним странам')
    plt.xlabel('Год')
    plt.ylabel('Индекс счастья') 
    plt.legend(t_df, bbox_to_anchor=(1, 0.6))
```

Наша страна по уровню счастья по сравнению с соседями занимает срединное положение. Наивысший пик счастья зафиксирован в России в 2017 году, после чего индекс счастья стал снижаться. По сравнению с 2017 годом мы пропустили вперёд Японию и Казахстан, однако стали более счастливыми по сравнению с Белоруссией.   
Самой счастливой из анализируемых стран-соседей России является Финляндия, самой несчастной - Украина. Уровень счастья из стран-соседей России на протяжении 5 лет с 2015 по 2019 годов устойчиво растёт в Финляндии и Польше, устойчиво падает в Белоруссии и США.  
В Японии и КНР уровень счастья отличается стабильностью.

Удивила высокая степень свободы выбора в КНР. Не удивили высокий уровень восприятия коррупции в России, Украине и Польше

```python
# Какие страны больше всего разочаровались и воодушевились жизнью за период между 2015 и 2019

All = df_2019.merge(df_2018, how='left', on='Country or region').rename({'Country or region': 'Country'}, axis=1).merge(df_2017, how='left', on='Country').merge(df_2016, how='left', on='Country').merge(df_2015, how='left', on='Country')

All_hapiness_score = All[['Country', 'Happiness Score_y','Happiness Score_x', 'Happiness.Score', 'Score_y', 'Score_x']].rename({'Happiness Score_y': '2015','Happiness Score_x': '2016', 'Happiness.Score': '2017', 'Score_y': '2018', 'Score_x': '2019'}, axis=1)
All_hapiness_score['dif'] = All_hapiness_score['2019'] - All_hapiness_score['2015']
All_hapiness_score.sort_values('dif', ascending=False)
```
