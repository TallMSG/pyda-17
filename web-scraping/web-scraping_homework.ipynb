```python
# Ответ по доработке

KEYWORDS = ['python', 'парсинг']
req = requests.get('https://habr.com/ru/all/')
soup = BeautifulSoup(req.text, 'html.parser')
posts = soup.find_all('article', class_='post')
articles_title = list(map(lambda x: x.find('a', class_='post__title_link'), posts))
articles = list(map(lambda x: x.find('div', class_='post__body post__body_crop'), posts))
hubs = list(map(lambda x: x.find('a', class_='inline-list__item-link hub-link'), posts))
times = list(map(lambda x: x.find('span', class_='post__time'), posts))

habr_news = pd.DataFrame()

for title, article, hub, time in zip(articles_title, articles, hubs, times):
    article = article.text.lower()
    title1 = title.text.lower()
    hub = hub.text.lower()
    search_results = [keyword in article or keyword in title1 or keyword in hub for keyword in KEYWORDS]
    if any(search_results):
        row = {'дата': time.text, 'заголовок': title.text, 'ссылка': title.attrs.get('href')}       
        habr_news = pd.concat([habr_news, pd.DataFrame([row])])
habr_news
```

# Задание 2

Написать скрипт, который будет проверять список e-mail адресов на утечку при помощи сервиса Avast Hack Ckeck. Список email-ов задаем переменной в начале кода:
EMAIL = [xxx@x.ru, yyy@y.com]

В итоге должен формироваться датафрейм со столбцами: <почта> - <дата утечки> - <источник утечки> - <описание утечки>

Подсказка: сервис работает при помощи "скрытого" API. Внимательно изучите post-запросы.

```python
EMAIL = ['xxx@x.ru', 'yyy@y.com']
```

```python
URL = 'https://identityprotection.avast.com/v1/web/query/site-breaches/unauthorized-data'

params = {'emailAddresses': ['xxx@x.ru', 'yyy@y.com']}
headers = {
    'Accept': 'application/json, text/plain, */*',
    'Accept-Encoding': 'gzip, deflate, br',
    'Accept-Language': 'ru-RU,ru;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Content-Length': '31',
    'Content-Type': 'application/json;charset=UTF-8',
    'Host': 'identityprotection.avast.com',
    'Origin': 'https://www.avast.com',
    'Referer': 'https://www.avast.com/',
    'sec-ch-ua': '"Chromium";v="88", "Google Chrome";v="88", ";Not A Brand";v="99"',
    'sec-ch-ua-mobile': '?0',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'same-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.190 Safari/537.36',
    'Vaar-Header-App-Build-Version': '1.0.0',
    'Vaar-Header-App-Product': 'hackcheck-web-avast',
    'Vaar-Header-App-Product-Name': 'hackcheck-web-avast',
    'Vaar-Version': '0'   
}
```

```python
req = requests.post(URL, json=params, headers=headers, verify=True)
req.json()
```

```python
# Извлекаем датафрейм с id утечек для конкретных адресов из тега summary

email_df = pd.DataFrame()
for el in EMAIL:
    for eli in req.json()['summary'][el]['breaches']:
        row1 = {'id утечки': eli, 'email': el}
        email_df = pd.concat([email_df, pd.DataFrame([row1])])
email_df
```

```python
# Извлекаем датафрейм с id, датой, источником и описанием утечек из тега breaches

breaches_df = pd.DataFrame()
for k, v in (req.json()['breaches']).items():
    row = {'id утечки': v['breachId'], 'дата утечки': v['publishDate'], 'источник утечки': v['site'], 'описание утечки': v['description']}
    breaches_df = pd.concat([breaches_df, pd.DataFrame([row])])
breaches_df
```

```python
# Объединяем датафреймы с емейлами и описанием утечек по id утечки

final_df = email_df.merge(breaches_df, how='left', on='id утечки').drop(['id утечки'], axis=1)
final_df
```
